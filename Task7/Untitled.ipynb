```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.animation as animation
```

```python
def init():
    line.set_data([], [])
    return line,
```

Задаем функцию по x и функцию по y для фигур Лиссажую

```python
def function_x(a_b):
    return 30 * np.sin(a_b * t)


def function_y():
    return 30 * np.sin(1 * t)
```

Добавляем функцию для анимации фигур Лиссажу. Увеличиваем отношение a/b

```python
def animate(i):
    a_b = -10 + 0.003 * i

    x = function_x(a_b)
    y = function_y()

    xdata, ydata = [], []
    xdata.append(x)
    ydata.append(y)

    line.set_data(xdata, ydata)
    return line,
```

Разлиновываем плоскость, задаем t, анимируем фигуры и показываем plot.

```python
if __name__ == "__main__":
    plt.style.use('dark_background')
    fig = plt.figure()
    ax = plt.axes(xlim=(-50, 50), ylim=(-50, 50))

    plt.title('Фигуры Лиссажу')
    plt.axis('off')

    line, = ax.plot([], [], lw=2)
    t = np.arange(-10, 10, 0.001)
    anim = animation.FuncAnimation(fig, animate, init_func=init, frames=5000, interval=1, blit=True)
    plt.show()
```
